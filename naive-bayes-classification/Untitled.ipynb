```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
X = df.iloc[:,:-1].values
```

```python
X
```

```python
pca = PCA()
X_2d = pca.fit_transform(X)
```

```python
X_2d
# df['X1'] = X_2d[:,0]
# df['X2'] = X_2d[:,1]
```

```python
sns.barplot(x=list(df.columns[:4]), y=pca.explained_variance_)
```

```python
sns.scatterplot(x= df['X1'], y=df['X2'], hue=df['species'])
```

```python
tsne_obj = TSNE(learning_rate=10)
```

```python
X_2d_2 = tsne_obj.fit_transform(X)
df['X1_TSNE'] = X_2d_2[:,0]
df['X2_TSNE'] = X_2d_2[:,1]
```

```python
sns.scatterplot(x= X_2d_2[:,0], y=X_2d_2[:,1])
```

```python
sns.scatterplot(x= df['X1_TSNE'], y=df['X2_TSNE'], hue=df['species'])
```
